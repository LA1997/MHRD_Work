```python
keydict = {}
```

```python
for i in range(len(df)):
    s1 = df.loc[i,'DE']
    if(type(s1)==float):
        df.drop(index=i,axis=0,inplace=True)
    else:
        s1 = s1.split('; ')
        for j in s1:
            if(j in keydict.keys()):
                keydict[j].append(i)
            else:
                keydict[j]=[i]
            
```

```python
skeydict = sorted(keydict.items(), key = lambda kv:(len(kv[1]), len(kv[0])))
skeydict.reverse()
```

```python
lkeys = []
for i in skeydict:
    lkeys.append(i[0])
```

```python
len(lkeys)
```

```python
#Merge Similar Words
import difflib
simword = {}
i = 0
while(i<len(lkeys)):
    simword[lkeys[i]] = list()
    j=i+1
    while(j<len(lkeys)):
        seq = difflib.SequenceMatcher(None,lkeys[i],lkeys[j])
        d = seq.ratio()*100
        if(d>50):
            simword[lkeys[i]].append(lkeys[j])
            keydict[lkeys[i]] = keydict[lkeys[i]]+keydict[lkeys[j]]
            del keydict[lkeys[j]]
            lkeys.remove(lkeys[j])
            
        j+=1
    i+=1
    print(i)
            

    
```

```python
len(keydict)
```

```python
def sortdict(dt):
    tup = sorted(dt.items(), key = lambda kv:(kv[1], kv[0]))
    tup.reverse()
    return tup
```
