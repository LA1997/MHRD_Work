```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('KeywordinCitationsReduced.xlsx')
```

```python
df
```

```python
def stringtolist(li):
    return [x.strip("'") for x in li.strip('][').split(', ')]

for i in range(len(df)):
    df = df.astype(object)
    df.loc[i,'Count'] = stringtolist(df.loc[i,'Count'])
    df.loc[i,'Total'] = len(df.loc[i,'Count'])
    print(i)
```

```python

# 2019 Division 
labels = list(df['Keyword'])
sizes = list(df['2019'])

colors = []
explode = (100, 0, 100, 0,100,0,100,0,0) # explode 1st slice


fig1, ax1 = plt.subplots(figsize=(6, 5))
fig1.subplots_adjust(0.3,0,1,1)


# Plot
plt.pie(sizes, explode=explode,autopct='%1.1f%%', shadow=True, startangle=140,radius=1000)

ax1.axis('equal')

total = sum(sizes)
plt.legend(
    loc='upper right',
    labels=['%s, %1.1f%%' % (
        l, (float(s) / total) * 100) for l, s in zip(labels, sizes)],
    prop={'size': 11},
    bbox_to_anchor=(0.0, 1),
    bbox_transform=fig1.transFigure
)

plt.show()
```

```python
# Overall Division 
labels = list(df['Keyword'])
sizes = list(df['Total'])

colors = []
explode = (100, 0, 100, 0,100,0,100,0,0)  # explode 1st slice


fig1, ax1 = plt.subplots(figsize=(6, 5))
fig1.subplots_adjust(0.3,0,1,1)


# Plot
plt.pie(sizes, explode=explode,autopct='%1.1f%%', shadow=True, startangle=140,radius=1000)

ax1.axis('equal')

total = sum(sizes)
plt.legend(
    loc='upper right',
    labels=['%s, %1.1f%%' % (
        l, (float(s) / total) * 100) for l, s in zip(labels, sizes)],
    prop={'size': 11},
    bbox_to_anchor=(0.0, 1),
    bbox_transform=fig1.transFigure
)

plt.show()
```
